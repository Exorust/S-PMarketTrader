```python
import nest_asyncio
nest_asyncio.apply()
from ib_insync import *
ib = IB()
ib.connect('127.0.0.1', 7497, clientId=1)
```

```python
contract = Future('ES', '202306', 'CME')
```

```python
# Define the future contract with more details
contract = Future(
    symbol='ES',
    lastTradeDateOrContractMonth='202406',
    exchange='CME',  
)

# Check if the contract can be resolved
contract_details = ib.reqContractDetails(contract)
```

```python
from datetime import date
import pandas as pd
final_df = pd.DataFrame()
min_date = date.today()
end_date = ''
counter  = 1
while min_date>date(2023,3,17):
    print(f"Downloading 1 month {counter} time")
    historical_data = ib.reqHistoricalData(
        contract, endDateTime=min_date, durationStr='1 M',
        barSizeSetting='1 day', whatToShow='MIDPOINT', useRTH=True)
    df = util.df(historical_data)
    end_date = df['date'].max()
    min_date = df['date'].min()
    print(min_date)
    final_df = pd.concat([final_df, df], axis = 0)
    counter+=1
```

```python
final_df = final_df.drop_duplicates(subset = ['date'])
```
